```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    if 0 <= x < .5:
        return 2*x
    elif .5 <= x <= 1:
        return 2 - 2*x
    else:
        return None
```

```python
runs = 4 # number of runs
N = 100 # iteration length

for run in range(runs):
    
    x0 = np.random.rand()
    
    for precision in ['float32','float64']:
    
        x_iter = np.array([x0], dtype=precision)
        for i in range(N-1):
            x_new = np.array([f(x_iter[-1])], dtype=precision)
            x_iter = np.append(x_iter, x_new)
        
        plt.plot(range(N), x_iter)
        plt.title("x_0 = {}, precision = {}".format(str(x_iter[0]),precision))
        plt.xlabel("n")
        plt.ylabel("x_n",rotation=0)
        plt.savefig("hw1 run = {}, precision = {}".format(run,precision))
        plt.show()
```
