```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
nx = 101
ny = 101
L = 1
xlist = np.linspace(-L, L, nx)
ylist = np.linspace(-L, L, ny)
x, y = np.meshgrid(xlist, ylist)
zlist = [-.4,-.3,-.2,-.1,0,.1,.2,.3,.4]
a = 1
```

```python
for z0 in zlist:
    
    detA = np.zeros((ny,nx))
    nucA = np.zeros((ny,nx))
    for i in range(ny):
        for j in range(nx):
            x0 = x[i,j]
            y0 = y[i,j]
            A = np.array([[0.5,x0],[y0,z0]])
            detA[i,j] = la.det(A)
            nucA[i,j] = la.norm(A, ord='nuc')

    plt.contour(x, y, detA, [0], colors='red')
    plt.contour(x, y, nucA, [a])
    plt.title('z = {}'.format(z0))
    plt.xlabel('x')
    plt.ylabel('y', rotation=0)
    plt.savefig('hw7 level curves z = {}.png'.format(z0))
    plt.show()
```
