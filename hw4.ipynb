```python
import numpy as np
import numpy.random as rand
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
def RayleighQuotient(n):
    print('n = {}'.format(n))
    A = rand.randn(n,n)
    A = A.T + A
    v = rand.randn(n,1)
    v = v/la.norm(v)
    k = 0
    mu = np.array([], dtype=np.float64)
    mu = np.append(mu, v.T@A@v)
    tol = 1e-12
    I = np.eye(n)
    res = np.abs(la.norm(A@v - mu[k]*v)/mu[k])
    print('k = {}: lam = {} \t res = {}'.format(k,mu[k],res))
    while res > tol:
        w = la.solve(A - mu[k]*I, v)
        k = k + 1
        v = w/la.norm(w)
        mu = np.append(mu, v.T@A@v)
        res = np.abs(la.norm(A@v - mu[k]*v)/mu[k])
        print('k = {}: lam = {} \t res = {}'.format(k,mu[k],res))
    return mu
```

```python
for n in [10,100,1000]:
    for run in range(4):
        mu = RayleighQuotient(n)
        mu_k = mu[-1]
        err = np.abs(mu[:-1]-mu_k)

        plt.title('e_(k+1)/e_k^3 for n = {}, run = {}'.format(n,run))
        plt.xlabel('k')
        k = err.size
        plt.scatter(range(k-1), [err[j+1]/err[j]**3 for j in range(k-1)])
        plt.xticks(range(k-1))
        plt.savefig('hw4 err n = {} run = {}.png'.format(n,run))
        plt.show()
```
